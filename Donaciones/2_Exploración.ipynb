### Lectura de archivo parquet

## Análisis Exploratorio

Seleccionamos las columas de interes, que son las que nos podrán brindar información relevante.

```
cols = ["ID_CONTACTO","ESTRATO","MONTO_TOTAL","NIVEL_JERARQUICO","GENERO","EDAD","DONACIONES","OPORTUNIDADES"]
```

Creamos una vista SQL con la información de las columnas seleccionadas

```
donaciones.select(cols).createOrReplaceTempView("donaciones")
```

Conteo de registros

```
%sql
SELECT COUNT(*)  Registros
FROM donaciones
```

Cantidad de personas

```
%sql
SELECT COUNT(DISTINCT ID_CONTACTO) TotalPersonas FROM donaciones
```

Cantidad de donaciones

```
%sql
SELECT SUM(DONACIONES) TotalDonaciones FROM donaciones
```

Cantidad personas por Estrato

```
%sql
SELECT ESTRATO, COUNT(DISTINCT ID_CONTACTO) PersonasEstrato
FROM donaciones
GROUP BY ESTRATO
ORDER BY ESTRATO asc
```

Cantidad de donaciones por estrato

```
%sql
SELECT ESTRATO, SUM (DONACIONES) DonacionesEstrato
FROM donaciones
GROUP BY ESTRATO
ORDER BY ESTRATO asc
```

% de donaciones por Nivel Jerárquico

```
%sql
SELECT NIVEL_JERARQUICO, SUM (DONACIONES)/COUNT(ID_CONTACTO) DONACION
FROM donaciones
GROUP BY NIVEL_JERARQUICO
```

Monto donado por Edad

```
%sql
SELECT EDAD, SUM(MONTO_TOTAL)
FROM donaciones
GROUP BY EDAD
ORDER BY EDAD ASC
```

Promedio de apuesta por genero

```
%sql
SELECT ESTRATO, SUM(MONTO_TOTAL)/SUM (DONACIONES) PROMDONADO
FROM donaciones
GROUP BY ESTRATO
```
