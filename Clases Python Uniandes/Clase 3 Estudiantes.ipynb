3. Escriba una función que construya el [coeficiente de variación](https://es.wikipedia.org/wiki/Coeficiente_de_variaci%C3%B3n) de un vector/lista dado/a.

```python
def ListDivisores(numero):
    lista=[] # Creo un lista vacía a la que le voy a añadir los divisores
    for i in range(numero): # Haga un loop desde 0 hasta el número deseado
        if numero % (i+1)==0: # Si el residuo es cero entonces es múltiplo
            lista.append(i+1) # Hago i+1 y no i porque arranca desde cero y no desde 1. 
    if len(lista)==2: # Si solo hay dos divisores, significa que es el 1 y el número en si, luego es un primo. 
        print('El ' ,numero, ' es un número de primo')
    return(lista)
```

## punto 4: Hacer uno algoritmo que luego de ingresar 2 números por teclado, imprima los números naturales que hay entre ambos empezando por el más pequeño, contar cuantos números hay y cantos de ellos son pares.

```python
def wahio():
    num1=int(input('numero 1: '))
    num2=int(input('numero 2: '))
    if num1>num2:
        a=num2
        b=num1
    else:
        b=num2
        a=num1
    vector=[str(i) for i in range(a+1,b)]
    pares=[str(i) for i in range(a+1,b) if i%2==0]
    print(' Hay',len(vector),'numeros: ','-'.join(vector[0:10]),'....')
    print(' Hay',len(pares),'numeros pares: ','-'.join(pares[0:10]),'....')
seguir='SI'
while seguir=='SI':
    wahio()
    seguir=input('otro numero <SI - NO> ').upper()
```

## Punto 7: hacer un algoritmo que compruebe si un sudoku (terminado) está solucionado correctamente.

```python
ma=[['5', '3', '4', '6', '7', '8', '9', '1', '2'], ['6', '7', '2', '1', '9', '5', '3', '4', '8'], ['1', '9', '8', '3', '4', '2', '5', '6', '7'], ['8', '5', '9', '7', '6', '1', '4', '2', '3'], ['4', '2', '6', '8', '5', '3', '7', '9', '1'], ['7', '1', '3', '9', '2', '4', '8', '5', '6'], ['9', '6', '1', '5', '3', '7', '2', '8', '4'], ['2', '8', '7', '4', '1', '9', '6', '3', '5'], ['3', '4', '5', '2', '8', '6', '1', '7', '9']]
for i in ma:
    print(' | '.join(i))
    print('--'*17)
def verificacion():
    con1,con2=0,0
    vector=[]
    for a in range(9):
        for b in range(9):
            valor=ma[a][b]
            vector.append(valor)
        if len(set(vector))==9:
            con1+=1
            vector=[]
            pass
        else:
            print('hay un error en la fila',a+1)
            break
    for a in range(9):
        for b in range(9):
            valor=ma[b][a]
            vector.append(valor)
        if len(set(vector))==9:
            con2+=1
            vector=[]
            pass
        else:
            print('hay un error en la columna',a+1)
            break
    return con1,con2
r1,r2=verificacion()
if r1==9 and r2==9:
    print('El sudoko está solucionado efectivamente')
```
