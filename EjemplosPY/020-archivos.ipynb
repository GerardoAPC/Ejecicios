```python
def read():
    path = '/content/drive/MyDrive/Colab Notebooks/EjemplosPY/archivos/numbers.txt'
    numbers = []
    with open(path, "r",encoding="utf-8") as f: 
        for line in f:
            numbers.append(int(line))
    print(numbers)
```

```python
def write():
    path = '/content/drive/MyDrive/Colab Notebooks/EjemplosPY/archivos/names.txt'
    names = ["Facundo", "Gregorio", "Marta", "Susana", "Jose","Hector","vLADIMIR"]
    with open(path, "w", encoding="utf-8") as f:
        for name in names: 
            f.write(name)
            f.write("\n")
```

```python
def write2():
    path = '/content/drive/MyDrive/Colab Notebooks/EjemplosPY/archivos/names.txt'
    names = ["Luis", "jhon", "Gerardo"]
    with open(path, "a", encoding="utf-8") as f:
        for name in names: 
            f.write(name)
            f.write("\n")
```

```python
def read_names():
     path = '/content/drive/MyDrive/Colab Notebooks/EjemplosPY/archivos/names.txt'
     with open(path, "r",encoding="utf-8") as f: 
        for line in f:
            print(line)
```

```python
def run():
   read()
   write()
   write2()
   read_names()
```
