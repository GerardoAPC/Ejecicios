# Assert Statements
Las afirmaciones son declaraciones que afirman o declaran un hecho con confianza en su programa. Por ejemplo, mientras escribe una función de división, está seguro de que el divisor no debe ser cero, afirma que el divisor no es igual a cero.
Las aserciones son simplemente expresiones booleanas que verifican si las condiciones son verdaderas o no. Si es cierto, el programa no hace nada y pasa a la siguiente línea de código. Sin embargo, si es falso, el programa se detiene y arroja un error

```python
def divisors(num):
    assert num >= 0, "Debes ingresar un número positivo"
    divisors = []
    for i in range(1, num + 1):
        if num % i == 0:
            divisors.append(i)
    return divisors
```

```python
def run():
    num = input('Ingresa un número: ')
    assert num.replace("-", "").isnumeric(), "Debes ingresar un número"
    print(divisors(int(num)))
    print("Terminó mi programa")
```

```python
# Punto de arranque 
run()
```
